## Calculate the response time using start and end date

```python
df=pd.read_csv("small_closed_incident_Drop_NA.csv",low_memory=False)
ending_time=pd.to_datetime(df["End_Date"])
ending_time
```

```python
starting_time=pd.to_datetime(df["Start_Date"])
starting_time
```

```python
time_diff_in_hour= (ending_time - starting_time).astype('timedelta64[h]')
time_diff_in_hour
```

```python
#df["time_diff_in_hour"]=time_diff_in_hour
df["time_diff_in_hour"]=time_diff_in_hour
df_with_positive_time_diff=df[df["time_diff_in_hour"]>=0]
df_with_positive_time_diff
```

```python
df_with_positive_time_diff.to_csv("small_closed_incident_drop_na_time_diff_positive.csv",index=False)
```

## Creating a month column corresponding to the starting date

```python
df_with_positive_time_diff["Month_number"]=ending_time.dt.month #Based on starting time
df_with_positive_time_diff
```
